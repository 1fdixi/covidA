```python
Q1=covida['rh'].quantile (0.25)

Q3=covida['rh'].quantile (0.75)

QIR=Q3-Q1

QIR

print("shape of data(before)",covida.shape )

covida=covida[~((covida["rh"]<(Q1-1.5*QIR))| (covida["rh"]>(Q3+1.5*QIR)))]  

print("shape of data(after)",covida.shape )
```
